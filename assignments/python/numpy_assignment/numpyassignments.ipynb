### Assignment 3: Array Operations

1. Create two NumPy arrays of shape (3, 4) filled with random integers. Perform element-wise addition, subtraction, multiplication, and division.
2. Create a NumPy array of shape (4, 4) with values from 1 to 16. Compute the row-wise and column-wise sum.

```python
arr3 = np.random.randint(1,12, size=(3,4))
arr3_1 = np.random.randint(1,12, size=(3,4))

print("Original First Array")
print(arr3)

print("Original Second Array")
print(arr3_1)

addition = arr3 + arr3_1
subtraction = arr3 - arr3_1
multiplication = arr3 * arr3_1
division = arr3 / arr3_1

print("Addition")
print(addition)
print("Subtraction")
print(subtraction)
print("Multiplication")
print(multiplication)
print("Division")
print(division)
```

```python
arr3_2 = np.random.randint(1,16,size=(4,4))

print("First Array:")
print(arr3_2)

print("Row wise sum: ")

for i in range(len(arr3_2)):
    row_wise_sum = 0
    for j in range(len(arr3_2[0])):
        row_wise_sum += arr3_2[i][j]
    print(f"Sum of {i}th Row : {row_wise_sum}")

print("Column wise sum: ")

for i in range(len(arr3_2[0])):
    col_wise_sum = 0
    for j in range(len(arr3_2)):
        col_wise_sum += arr3_2[j][i]
    print(f"Sum of {i}th Column : {col_wise_sum}")
```

### Assignment 4: Statistical Operations

1. Create a NumPy array of shape (5, 5) filled with random integers. Compute the mean, median, standard deviation, and variance of the array.
2. Create a NumPy array of shape (3, 3) with values from 1 to 9. Normalize the array (i.e., scale the values to have a mean of 0 and a standard deviation of 1).

```python
arr4_1 = np.random.randint(1,25, size=(5,5))
print(arr4_1)
print(np.mean(arr4_1))
print(np.median(arr4_1))
print(np.std(arr4_1))
print(np.var(arr4_1))
```

```python
arr4_2 = np.arange(1,10).reshape((3,3))
print(arr4_2)
arr4_min = np.min(arr4_2)
arr4_max = np.max(arr4_2)

arr4_mean = np.mean(arr4_2)
arr4_std = np.std(arr4_2)
normalized_array = (arr4_2 - arr4_mean) / arr4_std
print("Normalized Array")
print(normalized_array)
print(np.mean(normalized_array))
```

### Assignment 6: Linear Algebra

1. Create a NumPy array of shape (3, 3) representing a matrix. Compute its determinant, inverse, and eigenvalues.
2. Create two NumPy arrays of shape (2, 3) and (3, 2). Perform matrix multiplication on these arrays.

```python
arr6_1 = np.random.randint(1,10, size=(3,3))
print("Original array")
print(arr6_1)
print("Determinant")
print(np.linalg.det(arr6_1))
print("Inverse")
print(np.linalg.inv(arr6_1))
print("Eigenvalues")
print(np.linalg.eigvals(arr6_1))
```

### Assignment 7: Advanced Array Manipulation

1. Create a NumPy array of shape (3, 3) with values from 1 to 9. Reshape the array to shape (1, 9) and then to shape (9, 1).
2. Create a NumPy array of shape (5, 5) filled with random integers. Flatten the array and then reshape it back to (5, 5).

```python
arr7_1 = np.arange(1,10).reshape((3,3))
print("original array reshape(3,3)")
print(arr7_1)
print("reshaped array to (1,9)")
print(arr7_1.reshape(1,9))
print("reshaped again to (9,1)")
print(arr7_1.reshape(9,1))
```

```python
arr7_2 = np.arange(1,26).reshape(5,5)
print("Original Array")
print(arr7_2)
print("Flattened Array")
arr7_flattend = arr7_2.flatten()
print(arr7_flattend)
print("Reshaped back to (5,5)")
arr7_reshaped = arr7_flattend.reshape(5,5)
print(arr7_reshaped)
```
