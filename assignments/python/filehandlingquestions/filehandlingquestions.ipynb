```python
count = 0
with open("sample.txt","r") as file:
    content = file.read()
    temp = ""
    for letter in content:
        if letter is not " ":
            temp += letter
        else:
            count += 1
            temp = ""
    if temp !="":
        count += 1
print("Number of words are:", count)
```

```python
with open("data.txt","r") as file:
    content = file.read()
    content = content.replace("rishabh","Rishabh Singh")
    with open("data.txt","w") as file2:
        file2.write(content) 
```

```python
with open("reverse_sample.txt","r") as file:
    content = file.read()
    list_content = content.splitlines()
    reversed = ""
    for con in list_content:
        reversed += con[::-1] + "\n"
    with open("reversed_output.txt","w") as file2:
        file2.write(reversed)
```

```python
with open("sample.txt","r") as file:
    content = file.read()
    lines = content.splitlines()
    total_lines = len(lines)
    temp = ""
    count_char = 0
    count_words = 0
    for letter in content:
        if letter == " ":
            temp = ""
            count_words += 1
        elif letter in ['!', '"', '#', '$', '%', '&', "'", '(', ')', '*', '+', ',', '-', '.', '/', ':', ';', '<', '=', '>', '?', '@', '[', '\\', ']', '^', '_', '`', '{', '|', '}', '~']:
            count_char += 1
        else:
            temp+=letter

    print(f"word count : {count_words}, char count : {count_char}, total_lines : {total_lines}")            
            
```
